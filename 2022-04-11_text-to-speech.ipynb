```python
import PyPDF2
import fitz
```

```python
path = '/mnt/d/books/economics/idea/Federičová, M., & Pertold, F. (2022). Odchody z učitelské profese v Evropě.pdf'
```

```python
with fitz.open(path) as doc:
    text = ''
    for page in doc:
        text += page.get_text()
```

```python
text = text.replace('\n', ' ').replace('  ', ' ')
```

```python
text
```

```python
pdfReader = PyPDF2.PdfFileReader(path)
```

```python
all_text = ''
num_pages = pdfReader.getNumPages()

for p in range(num_pages):
    page = pdfReader.getPage(p)
    text = page.extractText()
    all_text += text
```
